# 크롤링

```python
from selenium import webdriver
#검색어 : 고혈압 - 네이버 뉴스 링크 / 검색어 : 당뇨 또한 링크만 바꾸어 반복
url = "https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%EA%B3%A0%ED%98%88%EC%95%95"
driver = webdriver.Chrome()
driver.get(url)
```

```python
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# 뉴스 모두 로드 - 스크롤

xp = '/html/body'
body = driver.find_element(By.XPATH, xp)

for i in range(150) :
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
 
```

```python
# 뉴스 헤드라인

css = '.sds-comps-text-type-headline1'
a = driver.find_elements(By.CSS_SELECTOR, css)
title = []
for i in range(len(a)) :
    t = a[i].text
    title.append(t)
title
```

```python
# 뉴스 내용

css = '.sds-comps-text-type-body1'
a = driver.find_elements(By.CSS_SELECTOR, css)
body = []
for i in range(len(a)) :
    t = a[i].text
    body.append(t)
body
```

```python
news = pd.DataFrame({"headline" : title, 'body' : body})
news.to_csv("news_고혈압.csv", index=False)
news.head()
```

```python
#텍스트 전처리

import re 
stopwords = ['고혈압', '물론','통해','이유는','이유?','환자','질환']

def cleaning(text) :
    text = re.sub(r'[^\w\s]',' ',text)
    text = re.sub(r'[\n\t]',' ',text)
    text = re.sub(r'\s+',' ',text)
    text = text.strip()

    text = ' '.join([word for word in text.split() if word not in stopwords])
    text = ' '.join([word for word in text.split() if len(word) > 1])

    return text

clean =news['body'].apply(cleaning)
clean
```

```python
#텍스트 토큰화

from mecab import MeCab

mecab = MeCab()
token = []
for t in clean :
    kwords = mecab.pos(t)
    token.append(kwords)
```

```python
from mecab import MeCab

mecab = MeCab()
token = []
for t in clean :
    kwords = mecab.pos(t)
    token.append(kwords)


tag = ['NNP','NNG','NP','IC',"MAG","VA"]

text_list = []

for kwords in token :
    kwords = [w[0] for w in kwords if w[1] in tag]
    rnew_text = ' '.join(kwords)
    text_list.append(rnew_text)

news['clean']=pd.DataFrame(text_list)
news
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

cvec = CountVectorizer(max_df=0.9, min_df=20, analyzer = 'word')
X = cvec.fit_transform(news['clean'])
x = X.toarray()
x.shape
```

```python
# 토큰 목록을 배열로 가져옴
names = cvec.get_feature_names_out()

names
```

```python
# 각 토큰의 빈도 총합
xsum = np.sum(x, axis=0)

# 빈도 총합으로 정렬(순서)
rank_ind = np.argsort(xsum)

vocab = cvec.vocabulary_
```

```python
# top30 키워드 추출
most_n = []
most_v = []
for ind in rank_ind[-30:]:
    most_n.append(names[ind]) # 토큰과 개수를 순서별로 저장
    most_v.append(xsum[ind])
print(most_n)
print(most_v)
```

```python
import matplotlib.pyplot as plt

# 색상
plt.figure(figsize=(6,7))
fcolor = '#212121'
gcolor = '#616161'

# 그래프 그리기
ax = plt.subplot()
hbars = ax.barh(most_n, most_v, height=0.7, color=gcolor)
ax.set_title('(검색어 : 고혈압) 빈도그래프', fontsize=20)

# 눈금 라벨설정
ax.tick_params(axis='both', labelsize=10, labelcolor=fcolor)

# 테두리 삭제
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# 그래프에 값 라벨 표현
ax.bar_label(hbars, fmt='%g', padding=8, color=gcolor, fontsize=10)

plt.savefig("검색어_고혈압_빈도그래프.png")
plt.show()
```
